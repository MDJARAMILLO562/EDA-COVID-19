#**Análisis exploratorio de la evolución e impacto de Covid-19 en los estados de Estados Unidos durante el 2020.**


##**Descripción del problema**

La enfermedad por coronavirus Covid-19 dejó miles de fallecidos en sus años más críticos. Su rápida propagación mediante gotas que se expulsan al hablar, estornudar toser o respirar, hizo que el contagio fuera muy fácil. Esta situación tuvo un impacto enorme en la salud pública de Estados Unidos. Aún cuando se tomaron medidas de confinamiento y campañas de vacunación, se registraron altos casos de mortalidad.

Diversas fuentes coinciden en que la evolución e impacto de Covid-19 en EE.UU fue significativa, en 2020 se presentó un aumento de casos positivos y muertes, en marzo de ese año Estados Unidos se convirtió en el país con más casos de Covid-19, superando a china, y en abril se convirtió en el país con más muertes en el mundo.

A partir de esto, resulta interesante analizar la evolución e impacto del covid-19 a nivel de estados, logrando identificar que estados fueron los más afectados y en que meses, medir indicadores como la capacidad hospitalaria y presión sanitaria con el fin de entender si juegan un papel importante en los desenlaces fatales.


##**Hipótesis**

Los estados con mayores tasas de positivos y hospitalizaciones, tienden a presentar un mayor número de muertes. Asimismo, esos estados presentan una baja capacidad hospitalaria y una alta presión sanitaria, condiciones que pudieron influir en la alta tasa de desenlaces fatales.

##**Tipo de problema**

Este es un problema de análisis exploratorio de datos. No se busca establecer causalidad, sino describir, visualizar y resumir la información disponible para identificar patrones y relaciones. Se trata, por tanto, de un análisis de tipo descriptivo y comparativo.


##**Variables de entrada**

- date: Corresponde a la fecha en la que se registraron los datos. Es esencial para analizar la evolución temportal del Covid-19, indentificando tendencias y picos claves. (Dataset_1)

- positive: casos positivos confirmados. Me permitrá evaluar la magnitud del contagio en cada estado.

- hospitalizedCurrently: Número de pacientes hospitalizados en el momento del reporte. Me va a permitir medir la presión sobre el sistema de salud y analizar su relación con el aumento de muertes. (Dataset_1)

- death: Número total de muertes. Es clave para cuantificar el impacto del virus (Dataset_1)

- population: Población del estado, utilizada para calcular tasas por cada 100,000 habitantes. Permitiendo hacer comparaciones equitativas(Dataset_2)

- inpatient_beds: Número total de camas reportadas por fecha. Permitirá establecer la presión y la capacidad hospitalaria de cada estado. (Dataset_3)


##**Variables objetivo**

Tasa de positivos por 100,000 hab

$$
\text{tasa_positivos_100k} = \left( \dfrac{\text{Incre_positive}}{\text{population}} \right) \times 100000
$$


Tasa de hospitalizaciones por 100,000 hab</br>
$$
\text{tasa_hospitalmedian_100k} = \left( \dfrac{\text{hospitalized_median}}{\text{population}} \right) \times 100000
$$

Tasa de muertes por 100,000 hab

$$
\text{tasa_muertes_100k} = \left( \dfrac{\text{Incre_death}}{\text{population}} \right) \times 100000
$$


Capacidad relativa</br>

$$
\text{capacidad_relativa_100k} = \left( \dfrac{\text{inpatient_beds_median}}{\text{population}} \right) \times 100000)
$$

Presión hospitalaria
$$
\text{presion_hospitalaria} = \left( \dfrac{\text{hospitalized_max}}{\text{inpatient_beds_max}} \right) \times 100)
$$

###**4.2.2- Gráfico de barras**

**Figura 2**

*Análisis de las variables positive, hospitalizedcurrently, death gráficos de barras*

```python
# Se hace una copia del df original
df_temp = df1.copy()
#se ordena por fecha los datos
df_temp = df_temp.sort_values("date")
#Para las variables acumuladas toma el ultimo valor del año por estado
acumulados = df_temp.drop_duplicates(subset="state", keep="last")[["state", "positive", "death"]]
#Agrupar los registros por estado, calcular la mediana por estado
hospital = df_temp.groupby("state", as_index=False)["hospitalizedCurrently"].median()

# Unir ambos resultados
resumen = pd.merge(acumulados, hospital, on="state", how="left")

# Estilo general
sns.set(style="whitegrid")

# Se usa el for para graficar cada variable individualmente y no repetir código
for var in ["positive", "hospitalizedCurrently", "death"]:
    plt.figure(figsize=(16, 6))
    sns.barplot(data=resumen, x="state", y=var, palette="crest", hue="state", legend=False)
    plt.title(f"{var.capitalize()} por estado", fontsize=15, fontweight="bold")
    plt.xlabel("Estado")
    plt.ylabel(var.capitalize())
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='plain', axis='y')  # Evita notación científica
    plt.tight_layout()
    plt.show()
```

**Análisis de gráfico de barras**

Como parte del análisis exploratorio inicial se gráfica las variables númericas positive, hospitalizedCurrently y death, mediante gráficos de barras. Esto permite observar el total de registros de casos positivos, muertes y la mediana de hospitalizados por estados en el año 2020.

A partir de esta revisión preliminar, se determinó que para analizar la evolución e impacto del Covid-19 durante el año 2020, es necesario agrupar los datos por mes, de modo que se pueda observar la dinámica temporal de los indicadores y su relación con la capacidad hospitalaria en los distintos estados.

En este sentido se identificó que las variables positive y death, son de tipo acumulativo, por lo cual se requiere calcular la tasa de incremento mensual que refleje el cambio entre meses consecutivos. Por otro lado, las variables hospitalizedCurrently e inpatient_beds registran valores diarios y presentan alta dispersión con valores atípicos notables, lo que indica que no siguen una distribución normal. En estos casos, la literatura estadística recomienda el uso de medidas de tendencia central robustas, como la mediana, para representar de manera adecuada el comportamiento de las variables. Ademas de que la media, seria arriesgado usarla porque esta se ve afectada por valores atípicos y estaría sobrestimando la carga y presión  hospitalaria típica.

En consecuencia, el siguiente paso consiste en:

1- Agrupar los datos del df1 y df3 por mes y estado </br>
2- Calcular las tasas de incremento mensual para variables acumulativas (positive - death) </br>
3- Obtener estadísticos representativos (mediana) en las variables hospitalizedcurrently e inpatient_beds </br>
4- unificar los 3 dataframes df1, df2, df3 mediante un proceso de merge.</br>
5- Cálculo de variables objetivo o indicadores.

Este proceso permitirá estandarizar y comparar entre estados y meses, facilitando el análisis descriptivo y la visualización de las tendencias que caracterizaron la evolución e impacto del covid-19 en los diferentes estados de Estados Unidos durante el 2020.

1- Agrupación del df1 por mes y estado </br>
2- Calcular las tasas de incremento mensual para variables acumulativas </br>
3- obtener estadísticos representativos en las variables hospitalizedcurrently e inpatient_beds (mediana)


```python
# Crear copia del dataframe original
df1_copy = df1.copy()

# Crear columna con año y mes
df1_copy['year_month'] = df1_copy['date'].dt.to_period('M')  # Convierte la fecha a formato mensual (2020-03)

# Asegurar que los datos estén ordenados por estado y fecha
df1_copy = df1_copy.sort_values(['state', 'date'])

#Agrupar por estado y mes
# Para acumulados (positive y death): tomar el último valor del mes
df1_monthly_accum = (
    df1_copy.groupby(['state', 'year_month'], observed=False)
    .last()[['positive', 'death']])

# Para hospitalizados (variable diaria): calcular mediana y max del mes
df1_monthly_hosp = (
    df1_copy.groupby(['state', 'year_month'], observed=False)
    .agg({'hospitalizedCurrently': ['max','median']}))

# Unir los dos resultados
df1_monthly = pd.concat([df1_monthly_accum, df1_monthly_hosp], axis=1).reset_index()

# Renombrar las columnas
df1_monthly.columns = [
    'state', 'year_month', 'positive_last', 'death_last',
    'hospitalized_median', 'hospitalized_max']

# Ordenar por estado y mes
df1_monthly = df1_monthly.sort_values(['state', 'year_month'])

# Cálculo de los incrementos mensuales
# Diferencia del valor acumulado entre un mes y el anterior dentro de cada estado
df1_monthly['Incre_positive'] = (
    df1_monthly.groupby('state')['positive_last'].diff().fillna(df1_monthly['positive_last']))
df1_monthly['Incre_death'] = (
    df1_monthly.groupby('state')['death_last'].diff().fillna(df1_monthly['death_last']))

# Mostrar las primeras filas
df1_monthly.head()
```

4- unificar los 3 dataframes df1, df2, df3 mediante un proceso de merge.

df1_monthly : dataframe 1 con los datos de Covid mensuales de positive, hospitalized_Currently, death, recovered. Se unira por las claves state, year_month

df2_copy: dataframe 2 con la información de la cantidad de población por estado. Se unira por la clave state

df3_monthly: Camas hospitalizaciones (inpatient_beds). Se unira por state, year_month

```python
# Paso 1: Unir df1_monthly con df3_monthly por state y year_month
df_merged = pd.merge(df1_monthly, df3_monthly, on=['state', 'year_month'], how='left')

# Paso 2: Unir el resultado con df2_copy por state
df_merged = pd.merge(df_merged, df2_copy, on='state', how='left')

# Paso 3: Revisar el resultado final
print(df_merged.shape)
df_merged.head()
```

5- Cálculos de las variables objetivo

Ahora bien, una vez unido el dataframe, se tiene el final df_merged que es con el que vamos a hacer los cálculos para obtener las variables objetivos y poder finalmente analizarlas y evaluar la hipotesis planteada.

Tasa de positivos por 100,000 hab

$$
\text{tasa_positivos_100k} = \left( \dfrac{\text{Incre_positive}}{\text{population}} \right) \times 100000
$$


Tasa de hospitalizaciones por 100,000 hab</br>
$$
\text{tasa_hospitalmedian_100k} = \left( \dfrac{\text{hospitalized_median}}{\text{population}} \right) \times 100000
$$

Tasa de muertes por 100,000 hab

$$
\text{tasa_muertes_100k} = \left( \dfrac{\text{Incre_death}}{\text{population}} \right) \times 100000
$$


Capacidad relativa</br>

$$
\text{capacidad_relativa_100k} = \left( \dfrac{\text{inpatient_beds_median}}{\text{population}} \right) \times 100000)
$$

Presión hospitalaria
$$
\text{presion_hospitalaria} = \left( \dfrac{\text{hospitalized_max}}{\text{inpatient_beds_max}} \right) \times 100)
$$





```python
# Calcular las variables objetivo
df_merged['tasa_positivos_100k'] = (df_merged['Incre_positive'] / df_merged['population']) * 100000
df_merged['tasa_hospitalmedian_100k'] = (df_merged['hospitalized_median'] / df_merged['population']) * 100000
df_merged['tasa_muertes_100k'] = (df_merged['Incre_death'] / df_merged['population']) * 100000
df_merged['capacidad_relativa_100k'] = (df_merged['inpatient_beds_median'] / df_merged['population']) * 100000
df_merged['presion_hospitalariamax'] = (df_merged['hospitalized_max'] / df_merged['inpatient_beds_max']) * 100
df_merged['presion_hospitalariamed'] = (df_merged['hospitalized_median'] / df_merged['inpatient_beds_median']) * 100

# Revisar que no haya divisiones por cero ni inf: ya que causan errores en los gráficos
#problemas en cálculos estadísticos, distorsión de resultados
#convierte cualquier valor inf en NaN
df_merged.replace([np.inf, -np.inf], np.nan, inplace=True)
```

Se  gráfica la evolución mes a mes de la tasa de positivos, tasa de hospitalizaciones y tasas de muertes por cada 100.000 hab en cada estado, resaltando los estados con las tasas más altas y los segundos con tasas elevadas en cada mes.

**Figura 11**

*Evolución mensual de la tasa de positivos, hospitalizaciones y muertes por COVID-19 por cada 100,000 habitantes en los estados de Estados Unidos durante 2020*

```python
#Configurar el panel del gráfico para que salga con cuadriculas y fondo blanco
sns.set(style="whitegrid")

#Crea una figura con 3 filas y 1 columna de subgraficos
#Usar subplots para gráficar varios plots de lineas y configurar el tamaño con figsize
fig, axes = plt.subplots(3, 1, figsize=(18, 20))

#Crear una copia del original para no modificar df_merged
df_line = df_merged.copy()

#Convertir el formato period de la fecha, a string para poder graficar ya que seaborn y matplotlib no reconocen period como formato en fecha
df_line['year_month_str'] = df_line['year_month'].astype(str)

# Filtrar solo desde marzo de 2020 en adelante ya que enero y febrero no tiene datos
df_line = df_line[df_line['year_month_str'] >= '2020-03']


#Se usa for para no repetir código y hacer los tres graficos asi interamos todo dentro de un for y el hace lo que le indiquemos
#para cada variable en cuestión
#lista de la Variables a graficar para iterar en el for y dibujar cada subplot
variables = [
    ("tasa_positivos_100k", "Tasa de positivos por 100k"),
    ("tasa_hospitalmedian_100k", "Tasa de hospitalizaciones por 100k"),
    ("tasa_muertes_100k", "Tasa de muertes por 100k")]


#Bucle principal: un subplot por cada variable col =son las tasas de la lista
#y title= Nombre de las variables en la lista
# zip() combina elementos de dos listas posición por posición
#Por cada eje del gráfico (ax) y cada par de valores (col, title) de la lista de variables,
#dibuja un gráfico con los datos de la columna col y pone el título title.

for ax, (col, title) in zip(axes, variables): #Definimos que variable va a usar cada ax
    # dentro del for definimos lo que va hacer para cada variable
    #eliminar las filas con NaN
    df_clean = df_line.dropna(subset=[col]) #hacemos una copia limpia del anterior df
    monthly_max = df_clean.copy()
    #para cada mes obtener el valor más alto de cada variable
    top_states = monthly_max.loc[monthly_max.groupby('year_month_str')[col].idxmax()]
    #obtener el segundo máximo por mes, .sort_values ordena los datos por mes en orden ascendente enero--diciembre
    #y dentro de cada mes pone el valor de la col en orden descendente del mayor a menor
    #Asi el valor más alto de cada mes, queda de primero y el segundo más alto queda después
    #.groupby agrupa por mes todos los de marzo, todos los de abril...
    #.nth(1) selecciona el segundo valor mas alto del mes
    #.reset_index():devuelve los grupos a formato normal (no como indice)
    second_states = (
        monthly_max.sort_values(['year_month_str', col], ascending=[True, False])
        .groupby('year_month_str')
        .nth(1)# selecciona la segunda fila (índice 1) de cada grupo
        .reset_index())

     # Gráfico de lineas todos los estados en gris
    sns.lineplot(
        data=monthly_max,#df con todos los estados con los valores de tasas por mes
        x='year_month_str',#eje x el mes
        y=col,#eje y la variable iterable
        hue='state',#color a cada estado
        palette=['lightgray'] * len(states),#colorea todas las lineas de estados de gris
        alpha=0.3,#transparencia de las lineas 1 totalmente opaco
        linewidth=1,#grosor de la linea
        legend=False,#que no aparezcan en la leyenda
        ax=ax)#hace que cada gráfico se dibuje en su propio espacio vertical

    # Gráfico de lineas para estados con valores más altos
    sns.lineplot(
        data=monthly_max[monthly_max['state'].isin(top_states['state'].unique())],#de todo el df monthly_max selecciona solo las filas cuyo state este dentro del df top_states
        x='year_month_str',#eje x los meses
        y=col,#eje y el title de la variable iterable
        hue='state',# color por estado
        palette=state_colors,#selecciona la lista de colores que se definio anteriormente
        linewidth=1.8,
        ax=ax)

     # Gráfico de lineas para estados con los segundos valores más altos
    sns.lineplot(
        data=monthly_max[monthly_max['state'].isin(second_states['state'].unique())],
        x='year_month_str',
        y=col,
        hue='state',
        palette=state_colors,
        linewidth=1.5,
        linestyle='--',
        legend=True,
        ax=ax)

     # Gráficar Puntos para saber exactamente los estados con máximos valores
    sns.scatterplot(
        data=top_states,
        x='year_month_str',
        y=col,
        hue='state',
        palette=state_colors,
        s=80,
        marker='o',
        legend=False,
        ax=ax)

    #Gráficar triangulos para segundos estados con valores máximos
    sns.scatterplot(
        data=second_states,
        x='year_month_str',
        y=col,
        hue='state',
        palette=state_colors,
        s=80,
        marker='^',
        legend=False,
        ax=ax)

    #Agregar las Etiquetas a los circulos y triangulos solo a los estados con valores más altos
    #Circulos: Estados con valores más altos
    for _, row in top_states.iterrows():
        ax.text(
            row['year_month_str'],
            row[col] + (row[col] * 0.07),
            f"{row['state']} ({int(row[col])})",
            fontsize=6,
            ha='center',
            fontweight='bold')

    #Triangulos: segundos estados con valores más altos
    for _, row in second_states.iterrows():
        ax.text(
            row['year_month_str'],
            row[col] - (row[col] * 0.07),
            f"{row['state']} ({int(row[col])})",
            fontsize=5,
            ha='left',
            fontweight='bold')

    # Personalización
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.set_xlabel("Mes", fontsize=10)
    ax.set_ylabel(title, fontsize=11)
    ax.tick_params(axis='x', rotation=45, labelsize=9)
    ax.tick_params(axis='y', labelsize=9)

     #Agregar la leyenda en cada gráfico
    ax.legend(
        title='Estados destacados',
        loc='center left',
        bbox_to_anchor=(1.02, 0.5),
        fontsize=6,
        title_fontsize=7,
        frameon=True)

# Ajustar los márgenes y espacios internos entre elementos
plt.tight_layout(pad=4)
#Ajustar espacio entre subgráficos
plt.subplots_adjust(hspace=0.35)
plt.show()
```

*Matriz de correlaciones y distribuciones de variables Covid-19*

```python
#Crear una copia del DataFrame original para no modificarlo
df_pair = df_merged.copy()

# Seleccionar solo las variables numéricas que quieres analizar
variables = [
    "tasa_positivos_100k",
    "tasa_hospitalmedian_100k",
    "tasa_muertes_100k",
    "capacidad_relativa_100k",
    "presion_hospitalariamed"]

#Eliminar filas con NaN en estas variables para evitar errores o gráficos incompletos
df_pair = df_pair[variables].dropna()

#Crear el pairplot
sns.set(style="whitegrid")
sns.pairplot(
    df_pair,
    diag_kind="kde",        # Muestra la distribución suavizada en la diagonal
    plot_kws={"alpha": 0.6, "s": 15, "edgecolor": "k", "color": "teal" },
    height=2.5) # Ajustes estéticos de los puntos

# Calcular matriz de correlación de Pearson
corr_matrix = df_pair.corr(method='pearson')
print(corr_matrix)

#Mostrar el gráfico
plt.suptitle("Matriz de relaciones entre variables COVID-19", fontsize=16, y=1.02)
plt.show()
```

Tasa de muertes_100k guarda una relación alta de 0.61 con presion_jospitalariamed, reforzando la idea de que los sistemas de salud sobrecargados presentan mayores riesgos de mortalidad.
